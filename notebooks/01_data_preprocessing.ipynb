```python
import cv2
import numpy as np

# Define image size
IMG_SIZE = (224, 224)

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    if image is None:
        return None

    # Resize or apply your custom preprocessing
    processed_image = cv2.resize(image, (224, 224))  # Example resizing
    
    # Ensure image is in uint8 format (CV_8U)
    if processed_image.dtype != 'uint8':
        processed_image = cv2.convertScaleAbs(processed_image)

    return processed_image
```

```python
dataset = os.getenv('DATASET_PATH')
print(f"Dataset path: {dataset}")
```

```python
speakers = []
for item in os.listdir(dataset):
    if os.path.isdir(os.path.join(dataset, item)):
        speakers.append(item)

print(f"Speakers: {speakers}")
print(f"No. of speakers is: {len(speakers)}")
```

```python
preprocessed_data = []
labels = []

for speaker in speakers:
    speaker_path = os.path.join(dataset, speaker)
    # for category in ['word', 'phrase']:
    for category in ['words']:
        category_path = os.path.join(speaker_path, category)
        
        for word_path in os.listdir(category_path):
            word = os.path.join(category_path, word_path)

            for iteration_folder in os.listdir(word):
                iteration_path = os.path.join(word, iteration_folder)

                for frame in os.listdir(iteration_path):
                    frame_path = os.path.join(iteration_path, frame)
                    if "color" in frame and frame.endswith('.jpg'):
                        processed_image = preprocess_image(frame_path)

                        if processed_image is not None:
                            preprocessed_data.append(processed_image)
                            labels.append(speaker)
                            print(f"Processed image: {frame_path}")

print(f"Preprocessed data: {len(preprocessed_data)}")
print(f"Labels: {len(labels)}")
```

```python
np.save('preprocessed_data.npy', preprocessed_data)
np.save('labels.npy', labels)
print("Data saved successfully!")
```
